```python
import sqlite3
import pandas as pd
```

# Creating the database

### Establishing a connection

```python
conn = sqlite3.connect("Trees.db") 
cur = conn.cursor()
```

### Reading in data

```python
df_trees = pd.read_csv(r'https://opendata.citywindsor.ca/Uploads/Trees%20in%20Right%20of%20Way.csv')
df_trees.head()
```

### Sending data to database

```python
df_trees.to_sql('trees', conn, if_exists='replace', index=False) 
```

```python
rows = cur.execute("SELECT * FROM trees").fetchall()
rows
```
